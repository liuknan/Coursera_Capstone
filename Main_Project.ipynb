# Segmenting and Clustering Neighborhoods in Toronto

```python
from bs4 import BeautifulSoup
import pandas as pd
import wikipedia
```

```python
web = wikipedia.page("List of postal codes of Canada: M").html()
df = pd.read_html(web,header=0)[0]
df = pd.DataFrame(df)
df = df[df.Borough != 'Not assigned']
df.reset_index(inplace=True, drop=True)
df.loc[df.Neighbourhood == 'Not assigned','Neighbourhood'] = df['Borough']
```

```python
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).to_frame()
df.reset_index(inplace=True)
print(df.shape)
df
```

```python
geo = pd.read_csv("Geospatial_Coordinates.csv")
geo.rename(columns = {"Postal Code":"Postcode"},inplace=True)
df = pd.merge(df,geo,on='Postcode')
df
```

```python
import folium
map = folium.Map(location=[43.67, -79.388141], zoom_start=12)
incidents = folium.map.FeatureGroup()
```

```python
df_toronto = df[df['Borough'].str.contains('Toronto')]
df_toronto = df_toronto[['Latitude','Longitude']]
for lat, lng, in zip(df_toronto.Latitude, df_toronto.Longitude):
    incidents.add_child(
        folium.features.Marker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        ).add_to(map)
    )
map
```
